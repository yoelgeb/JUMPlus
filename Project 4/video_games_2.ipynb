Group the sales data of each region on Year_of_Release and plot the same


```python
sales_year = data.groupby('Year_of_Release').sum()[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].sort_index()

sales_year_max = sales_year[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].max(axis=1)

sales_year.plot(kind='bar')
```

Group North American, Japan, European & Other region sales data as per Genre and plot it

```python
genre_data = data.groupby('Genre').sum()[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].sort_index()
genre_data
```

```python
ax = genre_data.plot(kind='bar', rot=55)
ax.set_ylabel('Total Numbers')
ax.set_xlabel('Genre Category')
```

Group North American, Japan, European & Other region sales data as per Platform and plot it

```python
platform_data = data.groupby('Platform').sum()[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].sort_index()
platform_data
```

```python
ax = platform_data.plot(kind='bar', rot=55)
ax.set_xlabel('Platform Genre')
```

Find out which Genre games are sold in highest numbers (create a countplot of seaborn which could do that)

```python
genre_sorted = data.sort_values(by=['Genre'])
sns.countplot(y=genre_sorted['Genre'])
```
